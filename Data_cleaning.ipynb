```python
df["salary estimate"]=df['salary estimate'].apply(lambda x : x.split("(")[0])
df["salary estimate"]=df['salary estimate'].apply(lambda x : x.replace('$',''))
df['hourly']=df['salary estimate'].apply(lambda x:1 if 'hr' in x else 0)
df["salary estimate"]=df['salary estimate'].apply(lambda x : float(x.split("/")[0].replace(',','')))



df.head()
```

```python
df['rating']=df['company'].apply(lambda  x: x[-3:] if '\n'in x else 'no rating')
df['company']=df['company'].apply(lambda x: x[:-4] if '\n'in x else x[:-1])


df.head()
```

```python
def title_extraction(title):
    if "data"in title.lower() and "analyst" in title.lower():
        return "data analyst"
    elif "data"in title.lower() and "scientist" in title.lower():
        return "data scientist" 
    elif "data"in title.lower() and "engineer" in title.lower():
        return "data engineer"
    elif "machine"in title.lower() and "learning" in title.lower():
        return "machine learning engineer"
    else:
        return "other"
        
def level_extraction(title="",description=""):
    se=title
    if se=="":
        se=description
    

    if 'sr' in se.lower() or "senior" in se.lower() or "lead" in se.lower() or "principal" in se.lower() or "manager" in se.lower():
        return "Senior"
    elif 'jr' in se.lower() or "junior" in se.lower() or "entry" in se.lower() or "associate" in se.lower() or "fresh" in se.lower():
        return "junior"
    else:
        return "mid"

    

    
```

```python
df['job_simpl']=df['job title'].apply(lambda x : title_extraction(x))
df['seniority']=df['job title'].apply(lambda x : level_extraction(title=x))
df['seniority']=df.apply(lambda x :level_extraction(description=x['job description']) if x.seniority=="mid" else x.seniority ,axis=1 )


df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python

print (stopwords.words('english') [0:50])
```

```python
def preprocess(text):
    text=text.lower()
    text=re.sub(r"[^a-zA-Z0-9]"," ",text)
    text=text.split(" ")
    while text.count(""):
      text.remove("")
    en_stops = set(stopwords.words('english'))
    for i in text:
      if i in en_stops:
        text.remove(i)

    return text 

```

```python
df["description_len"] = df['job description'].apply(lambda x:len(preprocess(x)))
df.head()
```
